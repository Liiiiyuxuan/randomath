#### <font color=yellow> Integral Curve </font>

A line that, at each of its points, is tangent to a vector field is called an ***integral curve*** of the direction field.

Let us denote the abscissa by the letter $t$ and the ordinate by the letter $x$. The slope of a line of the field is the known function $v(t)$, and an integral curve is the graph of an unknown function $\varphi$. The curve $x = \varphi(t)$ is an integral if and only if 
\begin{align*}
    \frac{d \varphi}{d x} \equiv v(t) 
    \qquad \textbf{ By Barrow's Thm: } \;\; \varphi = \int v \;dt + C
\end{align*}

Assume that the domain of definition of the function $\varphi$ is an interval $I$ of the $t$-axis. 

#### <font color=orange>Theorem</font>

A necessary and sufficient condition for the graph of a function $\varphi$ to be an integral curve is that the following relation hold for all $t$ in $I$:
\begin{align*}
    \frac{d \varphi}{d t} = v(t, \varphi(t))
\end{align*}

#### <font color=yellow>Solution</font>

The function $\varphi$ is called a ***solution*** of the differential equation 
\begin{align*}
    \dot{x} = v(t, x)
\end{align*}
if it satisfies the above relation. 

#### <font color=yellow>Initial Condition</font>

The solution $\varphi$ satisfies the ***initial condition*** $(t_0, x_0)$ if $\varphi(t_0) = x_0$.



#### <font color=green>Example</font> 

The solution of the simplest equation $\dot{x} = v(t)$ with initial condition $(t_0, x_0)$ is given by Barrow's formula:
\begin{align*}
    \varphi(t) = x_0 + \int_{t_0}^t v(\tau) \;d\tau
\end{align*}

**Proof:** From the Fundamental Theorem of Calculus, we know that 
\begin{align*}
    \varphi(t) - \varphi(t_0) = \int_{t_0}^t \frac{d \varphi}{d t} \; dt = \int_{t_0}^t v(\tau) \; d\tau
\end{align*}
so 
\begin{align*}
    \varphi(t) = \varphi(t_0) + \int_{t_0}^t v(\tau) \; d\tau = x_0 + \int_{t_0}^t v(\tau) \; d\tau
\end{align*}

### 1.5 One dimensional phase space

Consider the equation
\begin{align*}
    \dot{x} = v(x), \qquad x \in \mathbb{R}
\end{align*}
The above equation is <font color=yellow>_autonomous_</font> since right-hand side is independent of $t$.

```python
from IPython.display import Image
display(Image(filename='fig6.png', width=500, height=300))
```

The points where $v$ vanishes are called <font color=yellow>_equilibrium positions_</font> (also _stationary points_ or _singular points_) of the vector field. If $a$ is an equilibrium position, then $\varphi(t) = a$ is a solution of the equation. 

#### <font color=orange>Theorem</font>

The solution $x = \varphi(t)$ of the equation $\dot{x} = v(x)$ with a continuous nonvanishing right-hand side which satisfies the initial condition $(t_0, x_0)$ is given by formula
\begin{align*}
    t - t_0 = \int_{x_0}^x \frac{1}{v(\xi)} \;d\xi
\end{align*}
Conversely the function $x = \varphi(t)$ defined by the above formula is a solution and satisfies this initial condition.

## 2 Vector Fields on Line

### 2.1 Existence and Uniqueness of Solutions

Let $v$ be a smooth function defined on an interval $U$ of the real axis.

#### <font color=orange>Theorem</font>

The solution $\varphi$ of the equation $\dot{x} = v(x)$ with initial condition $(t_0, x_0)$ 

* exists for all $t_0 \in \mathbb{R}$ and $x_0 \in U$;

* is unique in the sense that any two solutions with the same initial condition coincide in some neighborhood of the point $t_0$;

* is given by Barrow's formula: 
\begin{align*}
    t - t_0 = \int_{x_0}^{\varphi(t)} \frac{1}{v(\xi)} \;d\xi \qquad & \text{ if } v(x_0) \neq 0 \\ 
    \varphi(t) \equiv x_0 \qquad & \text{ if } v(x_0) = 0 
\end{align*}

**Proof (from text book):** 

Suppose $c_0$ is not an equilibrium position. We have seen in section 1 that: 1) the solution is given by Barrow's formula in a neighborhood of the point $t_0$ and 2) the function $\varphi$ defined by this formula is a solution and satisfies the initial condition.

In the case when $x_0$o is an equilibrium position the function $\varphi(t) = x_0$ is also a solution, and the theorem is proved.
