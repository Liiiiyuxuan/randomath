```python
import math
import random

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 10*10, 10*200)
Ax1 = 0
Bx1 = 50
Ax2 = 0 
Bx2 = 18
Ax3 = 12 
Bx3 = 0 
Ax4 = 0 
Bx4 = 0
Ax5 = -14
Bx5 = 0

Ay1 = -60
By1 = -30
Ay2 = 0 
By2 = 8
Ay3 = 0 
By3 = -10
Ay4 = 0 
By4 = 0
```

```python
x = Ax1 * np.cos(t) + Bx1 * np.sin(t) + Ax2 * np.cos(2 * t) + Bx2 * np.sin(2 * t) + Ax3 * np.cos(3 * t) + Bx3 * np.sin(3 * t) + Ax4 * np.cos(4 * t) + Bx4 * np.sin(4 * t) + Ax5 * np.cos(5 * t) + Bx5 * np.sin(5 * t)

y = Ay1 * np.cos(t) + By1 * np.sin(t) + Ay2 * np.cos(2 * t) + By2 * np.sin(2 * t) + Ay3 * np.cos(3 * t) + By3 * np.sin(3 * t) + Ay4 * np.cos(4 * t) + By4 * np.sin(4 * t)

plt.plot(x, y, label="elephant")
    
plt.title(r'OUtline of an elephant')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend()
plt.show()
```
